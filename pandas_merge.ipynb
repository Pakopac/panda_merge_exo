## 1er Merge

```python
merge = pandas.merge(u_usage,u_device[['use_id','platform','device']], on='use_id')
merge.head()
```

```python
print(u_usage.shape)
print(u_device.shape)
print(merge.shape)
```

```python
#merge["use_id"].value_counts().to_frame()
#merge["use_id"].equals(u_usage['use_id']) 
merge['use_id'].isin(u_usage['use_id']).value_counts()
```

## Left Merge

```python
mergeLeft = pandas.merge(u_usage,u_device[['use_id','platform','device']], on='use_id', how="left")
```

```python
print(u_usage.shape)
print(mergeLeft.shape)
```

```python
mergeLeft.head()
```

## Right Merge

```python
rightMerge = pandas.merge(u_usage,u_device[['use_id','platform','device']], on='use_id', how="right")
```

```python
print(u_usage.shape)
print(rightMerge.shape)
```

## Outer Merge

```python
outerMerge = pandas.merge(u_usage,u_device[['use_id','platform','device']], on='use_id', how="outer")
```

```python
print("There are " + str(outerMerge.shape[0]) +  " unique values of use_id in our dataframes.")
print("Outer merge result has " + str(outerMerge.shape) +  " rows")
```
